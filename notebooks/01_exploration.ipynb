# Twitch Stream Analysis - Exploration

This notebook performs exploratory data analysis (EDA) on the **Top Twitch Channels (July 17 – Aug 16, 2025)** dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('../data/top_twitch_channels.csv')
df.head()
```

```python
# Basic info
df.info()
df.describe()
```

```python
# Feature Engineering: Custom Metrics
df['Viewers_per_Hour'] = df['Hours Watched'] / df['Airtime Hours']
df['Audience_Stability'] = df['Average Viewers'] / df['Peak Viewers']

df[['Channel', 'Viewers_per_Hour', 'Audience_Stability']].head()
```

```python
# Visualization: Airtime vs Average Viewers
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Airtime Hours', y='Average Viewers', size='Peak Viewers', legend=False, alpha=0.7)
plt.title('Airtime vs Average Viewers')
plt.show()
```
